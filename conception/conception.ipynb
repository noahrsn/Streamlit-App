###  MachineLearningProzess und Modelllogik
Das Herzstck von **Degrees of No Return** ist ein durchgngiger, transparenter MachineLearningWorkflow. Historische Temperatur, CO und Meeresspiegeldaten werden zunchst bereinigt, harmonisiert und zeitlich synchronisiert. Darauf aufbauend lernen Zeitreihenmodelle die Beziehung zwischen Emissionsentwicklung und lokaler Temperaturvernderung  nicht als Black Box, sondern nachvollziehbar und validierbar.

Die Modelle werden konsequent gegen historische Daten getestet, mit klaren Qualittsmetriken bewertet und iterativ verbessert. Ziel ist es, Temperaturprognosen mit einem **RMSE unter 0,2C** zu erreichen  ein Niveau, das fundierte Aussagen auf lokaler Skala erlaubt. Aus den projizierten Temperaturverlufen werden anschlieend direkt **Hitzetage** abgeleitet: eine Kennzahl, die fr Menschen unmittelbar verstndlich ist.

Parallel dazu werden globale Meeresspiegelprojektionen auf hochaufgelste digitale Hhenmodelle gelegt. So entstehen realistische, rumlich przise **berflutungsflchen**, die mit bestehenden Hochwassergefahrenkarten abgeglichen werden. Eine rumliche bereinstimmung von mindestens **85%** dient dabei als Qualittsziel.  
Das Ergebnis ist kein isoliertes Modell, sondern ein **integriertes Risikosystem**, das Temperatur und berflutungsgefahren gemeinsam betrachtet.
