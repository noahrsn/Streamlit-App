# Konzept 1 — Die Klima‑Risiko‑Zeitmaschine
**Thema:** Meeresspiegelanstieg & Extremwetter
---
### Forschungsfrage
> "Wie verändern sich das lokale Überflutungsrisiko und die Frequenz von Hitzetagen in [Region X] bis zum Jahr 2050 unter Berücksichtigung verschiedener CO2-Emissionsszenarien (RCP)?"
---
#### 1. Klare Problemstellung
Globale Klimadaten (z.B. "1.5 Grad Erwärmung") sind für Laien und lokale Entscheidungsträger zu abstrakt. Es fehlt ein Werkzeug, das komplexe wissenschaftliche Modelle in konkrete, lokale Risikoszenarien (z.B. "Mein Keller steht 2050 unter Wasser") übersetzt, ohne dass man Wissenschaftler sein muss.
#### 2. Zielgruppe
- **Kommunale Stadtplaner:** Zur Identifikation gefährdeter Infrastruktur.
- **Versicherungsanalysten:** Zur Risikobewertung von Immobilienportfolios.
- **Besorgte Immobilienbesitzer:** Zur persönlichen Risikoabschätzung.
#### 3. Erfolgsmetriken (Genauigkeit)
- **Zeitreihen-Prognose:** Ein RMSE (Root Mean Squared Error) von < 0.2°C bei der Temperaturvorhersage auf Testdaten.
- **Geo-Spatial Accuracy:** Eine Übereinstimmung der berechneten Überflutungsmaske von mindestens 85% im Vergleich zu offiziellen Hochwassergefahrenkarten (Ground Truth).
#### 4. Deploymentziel
- **Streamlit WebApp:** Interaktive Karte mit Time-Slider (1990–2050).
- **Key Feature:** Ein "Szenario-Switch" (Business as usual vs. Green Deal), der die Karte live neu einfärbt.
#### Datenquellen (Datasets)
- Temperatur: Kaggle: Earth Surface Temperature Data
- CO2-Daten: NOAA Mauna Loa CO2
- Meeresspiegel: NASA Sea Level Change Data
- Topographie (DEM): OpenTopography (SRTM Data)

# Konzept 2 — Der Solar‑Potenzial‑Detektor
**Thema:** Computer Vision & Energie
---
### Forschungsfrage
> "Welche Dachflächen in [Stadtviertel X] eignen sich aufgrund ihrer Geometrie und Ausrichtung für Photovoltaik und wie hoch ist ihr kumuliertes jährliches Stromerzeugungspotenzial?"
---
#### 1. Klare Problemstellung
Hausbesitzer nutzen ihre Dachflächen oft nicht für Photovoltaik, weil die initiale Einschätzung ("Lohnt sich das? Passt das auf mein Dach?") zu kompliziert ist oder teure Gutachten erfordert. Es fehlt eine niederschwellige, visuelle "Klick-und-Check"-Lösung für ganze Stadtviertel.
#### 2. Zielgruppe
- **Private Hausbesitzer:** Zur Senkung der Einstiegshürde.
- **Lokale Energieversorger (Stadtwerke):** Als Lead-Generator für Solar-Angebote.
- **Immobilien-Investoren:** Zur Bewertung des energetischen Potenzials.
#### 3. Erfolgsmetriken (Genauigkeit)
- **Segmentierung (U‑Net):** Ein IoU (Intersection over Union) Score von > 0.8. Das Modell muss die nutzbare Dachfläche zu 80% deckungsgleich mit der Realität erkennen (Abzug von Schornsteinen/Fenstern).
- **Ertragsprognose:** Der vorhergesagte Jahresertrag (kWh) darf maximal ±10% von Standard-PVGIS-Berechnungen abweichen.
#### 4. Deploymentziel
- **Streamlit WebApp:** Karte basierend auf Satellitenbildern.
- **Key Feature:** Klick auf ein Gebäude → Polygon legt sich um das Dach → Sidebar zeigt Investitionskosten und Amortisationszeit.
#### Datenquellen (Datasets)
- Gebäudeumrisse: OpenStreetMap (via Overpass Turbo) oder Microsoft Global Building Footprints
- Satellitenbilder: Mapbox Satellite API (oder lokale Open Data Geoportale).
- Sonneneinstrahlung: PVGIS Data (EU Science Hub)

# Konzept 3 — Der "Hidden Gem" Hunter
**Thema:** Tourismus & Anomaly Detection
---
### Forschungsfrage
> "Welche Orte in [Stadt X] weisen eine statistisch signifikante Anomalie zwischen extrem hoher Bewertungsqualität (Sentiment) und niedriger Bewertungsquantität (Popularität) auf?"
---
#### 1. Klare Problemstellung
Städtetourismus leidet unter "Overtourism" an wenigen Hotspots, während authentische Orte ("Hidden Gems") unentdeckt bleiben. Touristen finden auf Plattformen wie TripAdvisor oft nur die "Touristenfallen", da Standard-Algorithmen Popularität über intrinsische Qualität stellen.
#### 2. Zielgruppe
- **Individualreisende (Gen Z / Millennials):** Die authentische Erlebnisse abseits des Mainstreams suchen.
- **Destination Management Organizations (DMOs):** Die Besucherströme räumlich entzerren wollen.
- **Locals:** Die neue Orte in ihrer eigenen Stadt entdecken wollen.
#### 3. Erfolgsmetriken (Genauigkeit)
- **Anomalie-Erkennung:** Eine Precision@10 von > 70%. Von den Top 10 vorgeschlagenen "Geheimtipps" müssen 7 bei manueller Überprüfung tatsächlich "hoch bewertet" UND "wenig besucht" sein.
- **Sentiment-Analyse:** F1-Score von > 0.8 bei der Unterscheidung zwischen "authentisch ruhig" (positiv) und "verlassen/schlecht" (negativ).
#### 4. Deploymentziel
- **Streamlit WebApp:** Eine Karte, die initial fast leer ist. Punkte erscheinen erst nach Filterung.
- **Key Feature:** Ein "Abenteuer-Regler" (Slider), der bestimmt, wie "hidden" (und damit risikoreicher) die Vorschläge sein sollen.
#### Datenquellen (Datasets)
- Reviews & Ratings: Kaggle Google Maps/Yelp Datasets (Suche nach "Google Maps Store Reviews").
- POI Verifizierung: OpenStreetMap (OSM via Python osmnx)
